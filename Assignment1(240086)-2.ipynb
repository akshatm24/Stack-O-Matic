Part 3: Problem Solving with Python Basics (20 Points)

Learnt fibonacci sequence and how to use the for loop to write the code. Also learnt how to take user inputs and exception handling

```python
a=0
b=1
for i in range(100):
  if(i==0) :
    print(0)
  elif(i==1):
    print(1)
  else:
    num=a+b
    a=b
    b=num
    print(num)
```

```python
for num in range(1,50):
  if num % 3 == 0:
    print("Fizz")
  elif num % 5 == 0:
    print("Buzz")
  elif num % 3 == 0 and num % 5 == 0:
    print("FizzBuzz")
  else:
    print(num)
```

```python
inp = eval(input("Enter the list"))
n=len(inp)
inp.sort()
if n%2==0: #this is evem
  median1 = inp[n//2]
  median2 = inp[n//2 - 1] # n starting from 0
  median = (median1 + median2)/2 #average of two medians
else:
  median = inp[n//2]
print(median)
```
